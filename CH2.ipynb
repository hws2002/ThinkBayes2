# 2.3 베이즈 테이블
베이즈 갱신을 보다 편리하게 할 수 있는 도구이다.  

```python
import pandas as pd
table = pd.DataFrame(index = ['Bowl 1', 'Bowl 2'])
```

```python
table['prior'] = 1/2, 1/2
table
```

```python
table['likelihood'] = 3/4, 1/2
```

```python
table['unnorm'] = table['prior'] * table['likelihood']
table
```

```python
prob_data = table['unnorm'].sum()
prob_data
```

```python
table['posterior'] = table['unnorm'] / prob_data
table
```

# 2.4 주사위 문제

```python
table2 = pd.DataFrame(index = [6,8,12])
table2
```

```python
from fractions import Fraction
table2['prior'] = Fraction(1,3)
table2
```

```python
table2['likelihood'] = Fraction(1,6), Fraction(1,8), Fraction(1,12)
table2
```

```python
table2['unnorm'] = table2['prior'] * table2['likelihood']
table2
```

```python
prob_data = table2['unnorm'].sum()
prob_data
```

```python
table2['posterior'] = table2['unnorm'] / prob_data
table2
```

```python
def update(table):
  """ 사후확률을 계산함 """
  table['unnorm'] = table['prior'] * table['likelihood']
  prob_data = table['unnorm'].sum()
  table['posterior'] = table['unnorm'] / prob_data
  return prob_data
```

# 2.5 몬티 홀 문제

Xi : 차가 i번째 문에 있는 사건  
Ci : 맨 처음 i번째 문을 고르는 사건  
Gi : 몬티가 i번째 문을 여는 사건  
일때, 우리가 비교해야 하는 것은 P(X1|C1G3) 와 P(X2|C1G3)이다.  
이때 C1 = 1, C2 = 0, C3 = 0  임이 이미 주어졌으니(1번문을 이미 선택함)  
C는 제외하고 베이즈 테이블을 작성해도 좋다.

```python
table3 = pd.DataFrame(index= ['door1', 'door2', 'door2'])
table3['prior'] = Fraction(1,3)
table3
```

P( G3 | X1)을 생각해보자.  
차가 1번 문에 있다면, 우리가 이미 1번 문을 선택했으므로, 몬티는 2번 문과 3번 문 둘 중 하나를 열 수 있다. 그러므로 P(G3|X1) = 1/2 이다.  

반면 P(G3 | X2) 는 1임을 알 수 있다.  


```python
table3['likelihood'] = Fraction(1,2) , 1, 0
table3
```

```python
update(table3)
table3
```
